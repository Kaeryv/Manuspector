```python
from dataset import read_asc
from glob import glob
import matplotlib.pyplot as plt
from os.path import basename
import numpy as np
files = glob("./16_02_2024/*.asc")
plt.subplots(figsize=(12,12))
cmap = plt.colormaps.get("tab20")
for i, f in enumerate(files):
    wl, R = read_asc(f)
    plt.plot(wl, R, label=basename(f), c=cmap(i))
plt.axvline(1200)
plt.axis([1200,2300, 0, 100])
plt.legend()
```
